```python
df = pd.read_csv('data.csv')
df.head()
```

```python
df.iloc[2,:]
```

```python
print("Alley" , df["Alley"].notnull().sum())
print("Alley" , df["Alley"].isnull().sum())
```

```python
print("Sum of count" , df[(df['LotFrontage'] > 80.0) & (df['LotArea'] > 16000.0)]['Id'].count())
```

```python
df['is_rich'] = np.where(df['LotArea'] > 15000, 'rich', 'poor')
df['is_rich'].value_counts()
df.drop(columns=['is_rich','salary'], inplace=True)
```

```python
df.groupby('Neighborhood').agg({'LotArea': 'max'}).sort_values('LotArea', ascending=False).head
```

```python
df.groupby('Neighborhood').agg({
    'SalePrice': 'mean',
    'LotArea': 'max',
})[lambda x : x['LotArea'] > 10000].sort_values(by='SalePrice', ascending=False).head()
```

```python
df.groupby('LotConfig').agg({
    'Condition1': 'count',
    'OverallQual': 'mean'
}).sort_values(by='OverallQual', ascending=False).head()
```

```python
df['date'] = pd.to_datetime(df['YrSold'].astype(str) + '-' + df['MoSold'].astype(str))
df.set_index('Id', inplace=True)
df.head()
```

```python
df['total'] = df['MSSubClass'].apply(lambda x : x + df.iloc[x]['LotArea'])
df['total'].describe()
```

```python
df[(df['LotArea'] > 20000) & (df['LotFrontage'] < 200)].groupby('Neighborhood').agg({'LotArea': 'mean', 'LotFrontage': 'mean'}).head()
```

```python
df["Normalized"] = (df["LotArea"] - df["LotArea"].mean()) / df["LotArea"].std()
```

```python
df['total1'] = df.apply(lambda x : x['OverallQual'] + x['LotArea'] + x['LotFrontage'] , axis=1)
df.head()
```

```python
df['LotPercentage'] = df['LotArea'] / df['LotArea'].max() * 100

df.plot(kind='hist', x='LotPercentage', bins=20)
plt.show()
```
