```python
# Essential Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('./creditcard.csv')
df.head()
```

We can See that there is High difference in legit and fraudulent transactions :


```python
print("Amount of legit transactions:", round(df['Class'].value_counts()[0]/len(df['Class'])*100, 2), "%")
print("Amount of fraudulent transactions:", round(df['Class'].value_counts()[1]/len(df['Class'])*100, 2), "%")
```

```python
plot = df['Class'].value_counts().plot(kind='bar')
plot.set_title('Number of Fraudulent and Legit Transactions')
plot.set_xlabel('Class')
plot.set_ylabel('Number of Transactions')
plt.show()
```
